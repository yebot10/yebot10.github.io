####      Do opioid overdoses tend to be associated with less affluent areas —  that is, areas where families have lower incomes?

In order to answer this question I decided to use a scatter plot. This graph seemed like an effective tool in order to compared two numeric data fields, such as household income versus the opioid overdose rate.

```python
fig,ax = plt.subplots()
ax.scatter(opioidsVA['MedianHouseholdIncome'], opioidsVA['FPOO-Rate'], c = 'purple')
ax.xaxis.set_label_text('Median Household Income')
ax.yaxis.set_label_text('FPOO-Rate / Opioid OD per 100,000')
fig.suptitle('Does Median Household Income in Virginia Influence Opiod OD Rates?')
plt.show()
```

#### Some people who start with opioid addictions are reported to transition to heroin use. What is the relationship in Virginia counties between opioid overdoses and heroin overdoses? 

Similar to the question above, I thought a scatter plot was the most effective graph type for the data needed to show a possible correlation between the opioid overdose rates and the heroine overdose rates.

```python
fig,ax = plt.subplots()
ax.scatter(opioidsVA['FPOO-Rate'], opioidsVA['FFHO-Rate'], c = 'navy')
ax.xaxis.set_label_text('FPOO-Rate / Opioid OD per 100,000')
ax.yaxis.set_label_text('FFHO-Rate / Heroine OD per 100,000')
fig.suptitle('Do Opiod OD Rates in Virginia Influence Heroine OD Rates?')
plt.show()
```

#### Which states are associated with the greatest number of United States Presidents in terms of the Presidents’ birthplaces?

To display the data needed to answer the above question in the most concise and effective manner, I decided to employ the use of the bar graph. I also limited my bar graph to only contain the top 5 states since the question being asked only concerns which states birthed the "greatest number" of U.S. Presidents, not the total number for every state.

```python
PresStates = {}
for i in presidents['State']:
    if i in PresStates:
        PresStates[i] += 1
    else: 
        PresStates[i] = 1
PresStates = sorted(PresStates.items(), key = lambda president: president[1], reverse = True)
states,num_pres = zip(*PresStates)
```

```python
fig,ax = plt.subplots()
ax.bar(states[:5], num_pres[:5], color = 'maroon')
fig.suptitle('Which States Have Birthed the Most U.S. Presidents?')
ax.yaxis.set_label_text('Number of Presidents')
plt.xticks(rotation = 30)
```

#### How have vehicle sales in the United States varied over time?

For this final visualization exercise I decided on using a line graph / time series to show how the total number of cars sold each month had changed over time. I widened the horizontal size of the graph to help relieve some of the strain from having so many data points crunched together.

```python
fig,ax = plt.subplots(figsize=(20,4))
ax.plot(totalNSA['DATE'], totalNSA['TOTALNSA'])
ax.xaxis.set_label_text('Date | yyyy-mm-dd')
ax.yaxis.set_label_text('Number of Cars Sold (in Thousand Units)')
plt.xticks(rotation = 45)
plt.show()
```
